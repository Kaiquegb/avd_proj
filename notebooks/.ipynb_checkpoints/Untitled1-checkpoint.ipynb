```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Dados Simulados (Resultados do seu Clustering)
data = {
    'CODIGO_ESTACAO': ['A001_RECIFE', 'A002_PETROLINA', 'A003_CARUARU', 'A004_GARANHUNS', 'A005_ARARIPINA'],
    'CLUSTER_ID': [1, 2, 1, 1, 0],  # 1=Litoral, 2=Serto, 0=Agreste (exemplo)
    'LAT': [-8.0476, -9.3960, -8.2849, -8.8907, -7.5732],
    'LON': [-34.8770, -40.5033, -35.9696, -36.4958, -40.4969]
}
df_map = pd.DataFrame(data)

# 2. Configurar o Mapa
plt.figure(figsize=(10, 6))

# Cores e Legendas (Ajuste conforme seus resultados reais)
colors = {0: 'red', 1: 'blue', 2: 'green'}
labels = {0: 'Cluster 0 (Agreste)', 1: 'Cluster 1 (Litoral)', 2: 'Cluster 2 (Serto)'}

# 3. Desenhar
for cluster_id in df_map['CLUSTER_ID'].unique():
    subset = df_map[df_map['CLUSTER_ID'] == cluster_id]
    plt.scatter(subset['LON'], subset['LAT'], 
                c=colors.get(cluster_id, 'gray'), 
                label=labels.get(cluster_id),
                s=300, edgecolors='black', zorder=2)
    
    # Nomes das Cidades
    for _, row in subset.iterrows():
        plt.text(row['LON'] + 0.15, row['LAT'], row['CODIGO_ESTACAO'], fontsize=9, fontweight='bold')

# Esttica
plt.title('Tpico 7.4: Clusters Climticos de Pernambuco', fontsize=16)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title="Perfil Climtico")

# Mostrar
plt.tight_layout()
plt.show()
```
